### Create Test Data

```python

features_size=5 #number of features
def create_linear_data(n_samples=2000,f=features_size):   
    # Get regression data from scikit-learn
    x,y = datasets.make_regression(n_samples=n_samples,noise=20, n_features=f)
    x=pd.DataFrame(x)
    return x,y

x,y = create_linear_data()
# Vizualize the data
for i in range(0,features_size):
    plt.scatter(np.array(x[i]),y,alpha=0.5,label=f'Feature: {i+1}')
    plt.legend()
    plt.show()
```

### Compute Beta for every feature

```python
def __beta(x,y):
    '''
        Compute beta
    '''
    n = len(x)
    x_b = x.mean()
    y_b = y.mean()
    nume = 0  #numerator of the formula
    denom = 0 #denominator of the formula
    for i in range(0,n):
        nume += (x[i]-x_b)*(y[i]-y_b)  #summation of numerator
        denom += (x[i]-x_b)**2 #summation of denominator
    
    beta = nume/denom  #divide
    return beta

features = list(x.columns.values.tolist()) #get list of features
beta_l = []

for f in features:
    x_sub = x[f]
    beta = __beta(np.array(x_sub),y)
    beta_l.append(beta)

print(beta_l)
```

### Fit the line

```python
def __fit_line(x,y,b):
    f_l = list(x.columns.values.tolist()) #get list of features
    y_temp = []
    y_p = []
    for f in f_l:
        x_sub = x[f]
        y_sub = []
        for i in range(0,len(x_sub)):
            y_i = b[f]*x_sub[i]
            y_sub.append(y_i)
        y_temp.append(y_sub)
    for i in range(0,len(y)):
        _sum = 0
        for f in f_l:
            _sum+=y_temp[f][i]
        y_p.append(_sum)
        
    return x,y_p
    

x,y_p=__fit_line(x,y,beta_l)
#plot line fit
plt.plot(y, y_p, color = 'r') #plot the regression line only
plt.show()

```

### Evaluate using R Squared

```python
def get_rsquared(y,y_f):
    '''
        params: y = actual values, y_f = predicted values
    '''
    #compute total sum of squares
    ss_tot = 0
    y_b = y.mean()
    for i in range(0,len(y)):
        ss_tot+=(y[i]-y_b)**2
    #compute residual sum of squares
    ss_res = 0
    for i in range(0,len(y)):
        ss_res += (y[i]-y_f[i])**2
    
    return 1-(ss_res/ss_tot)

# get r squared for every feature
r_squared = get_rsquared(y,y_p)
print(f'R Squared: {r_squared}')
```

### Evaluate using RMSE

```python
def get_rmse(y,y_f):
    '''
        params: y = actual values, y_f = predicted values
    '''
    _se=0
    for i in range(0,len(y_f)):
        _se += (y[i]-y_f[i])**2
    rmse = math.sqrt(_se/len(y))
    return rmse

rmse = get_rmse(y,y_p)
print(f'RMSE: {rmse}')
```

### Evaluate using MSE

```python
def get_mse(y,y_f):
    '''
        params: y = actual values, y_f = predicted values
    '''
    _se = 0
    for i in range(0,len(y_f)):
        _se += (y[i]-y_f[i])**2
    mse = _se/len(y)
    return mse

mse = get_mse(y,y_p)
print(f'MSE: {mse}')

#using sklearn library
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y, y_p)
print(f'MSE sklearn: {mse}')
```
