```python
# import modules
import math
import pylab
import random
import statistics
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
# define helper functions

def __pdf(x,mu,sd):
    '''
        Computes the probability density function
    '''
    exp = math.exp(((x-mu)**2)/(2*(sd**2))*-1)
    pdf = (1/math.sqrt(2*math.pi*sd))*exp
    return pdf
```

### Normal distribution plots
    -Summary: create plot of normal distributions on different pdf computations,
            The greater the sd the wider and shorter distribution is created.
            If mu is changed to negative plot will skew to the left, if more positive
            it will skew to the right.

```python
def normal_distri_plot(num_range=100):

    data = [x/10.0 for x in range(num_range*-1,num_range)]  # create random distri variables
    plt.plot(data,[__pdf(x,0,0.5) for x in data],'-',label='mu=0,sd=0.5')
    plt.plot(data,[__pdf(x,0,1) for x in data],'--',label='mu=0,sd=1')
    plt.plot(data,[__pdf(x,0,2) for x in data],':',label='mu=0,sd=2')
    plt.plot(data,[__pdf(x,0,3) for x in data],':',label='mu=0,sd=3')
    plt.legend()
    plt.show()

normal_distri_plot()
```

### Normal probability qq plots
    -Summary: Create a qq plot the will compare distribution points on the line

```python
def normal_prob_qq(num_range=100):
    
    data = np.random.normal(0,1, num_range) #create random normal distributed data
    sm.qqplot(data, line='45')  #plot using sm module
    pylab.show()
    
normal_prob_qq()
```
