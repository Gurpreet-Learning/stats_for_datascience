```python
# import modules
import math
import pylab
import random
import statistics
import numpy as np
import pandas as pd
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
def __tstat(ms,sds,ns):
    '''
        let ms as list of mean, sds as list of sd and ns as list of n
    '''
    m = 0
    for i,mi in enumerate(ms):
        if i==0:
            m=mi
        else:
            m-=mi
    denom = 0
    for i,sdi in enumerate(sds):
        denom += sds[i]/ns[i]
    return m/math.sqrt(denom)

def __pdf(x,mu,sd):
    '''
        Computes the probability density function
    '''
    exp = math.exp(((x-mu)**2)/(2*(sd**2))*-1)
    pdf = (1/math.sqrt(2*math.pi*sd))*exp
    return pdf
```

```python
def t_distri(num_range=100):
    mu=0
    data = [x/10.0 for x in range(num_range*-1,num_range)]  # create random distri variables
    s_list = []
    t_list = []
    mu=0.5
    sd=0
    for i in range(2):
        _sample = [__pdf(x,sd,mu) for x in data]
        mu+=0.5
        s_list.append(_sample)
        
    
    ms = []
    sds=[]
    ns = []
    for _sample in s_list:
        _sample = np.array(_sample)
        ms.append(_sample.mean())
        sds.append(_sample.std())
        ns.append(len(_sample)) 
    t_val = __tstat(ms,sds,ns)
    p = 1 - stats.t.cdf(t_val,df=2)
    
    plt.plot(data,s_list[0],'-',label='mu=0,sd=0.5,t={}'.format(t_val))
    plt.plot(data,s_list[1],':',label='mu=0,sd=1,t={}'.format(t_val))
    plt.legend()
    plt.show()
    
    
t_distri()
```
